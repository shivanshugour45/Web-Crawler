```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('naukri_23-Jan-2023.csv')
```

```python
df.head()
```

```python
df.drop('job_id',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['min_exp'] = df['exp_req'].apply(lambda x: str(x).split('-')[0] if x != 'NA' else 'NA').astype('float64')
df['max_exp'] = df['exp_req'].apply(lambda x: str(x).split('-')[1].split(' ')[0] if x not in ['NA', np.nan] else 'NA')
```

```python
new_df = df.loc[(df['min_exp'] <= 3)]
```

```python
new_df.reset_index(drop = True, inplace = True)
```

```python
filtered = new_df.loc[df['role'].apply(lambda x: True if 'Data' in x or 'Analyst' in x or 'Science' in x or 'data' in x or 'analyst' in x else False)]
```

```python
filtered.reset_index(drop=True,inplace=True)
```
